Carga los datos

```python
#Se cargan los datos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df= pd.read_csv('Empresa Aliada.csv')
df
```

```python
df.info()
```

Identificar las variables clave para el análisis

```python
#Se crea grafica boxplot por categoria
fig, ax = plt.subplots()
sns.boxplot(data = df, y = 'TOTAL_UNIT_AVG_WEEKLY_SALES', x = 'CATEGORY')
plt.show()
```

```python
fig.savefig('Categoria.png')
```

Analizar la tendencia de ventas en el tiempo

```python
# Grafica de lineas de semanas y ventas totales
plt.plot(df['WEEK'], df['TOTAL_UNIT_AVG_WEEKLY_SALES'])
plt.xlabel('Semanas')
plt.ylabel('Productos')
plt.title('Tendencias de ventas')
plt.xticks(rotation=90)
plt.savefig('Ventas por semanas.png')
plt.show()
```

```python
# Grafica de lineas de región y ventas totales
plt.plot(df['REGION'], df['TOTAL_UNIT_AVG_WEEKLY_SALES'])
plt.xlabel('Región')
plt.ylabel('Ventas')
plt.title('Tendencias de ventas')
plt.xticks(rotation=90)
plt.savefig('Ventas por region.png')
plt.show()
```

```python
# Grafica de lineas de segmentos y ventas totales
plt.plot(df['SEGMENT'], df['TOTAL_UNIT_AVG_WEEKLY_SALES'])
plt.xlabel('Segmentos')
plt.ylabel('ventas')
plt.title('Tendencias de ventas')
plt.xticks(rotation=90)
plt.savefig('Ventas por segmento.png')
plt.show()
```

Explorar la relación entre diferentes variables

```python
#Se genera gráfica de dispersión (Scatter plots)
graf = sns.FacetGrid(df, col = 'SEGMENT')
graf.map(plt.scatter, 'ITEM', 'TOTAL_UNIT_AVG_WEEKLY_SALES')
plt.savefig('Relaciones entre variables clave.png')
```

Visualizar la distribución geográfica de las ventas

```python
#Se genera grafica de barras
sns.barplot(data = df, y = 'REGION', x = 'TOTAL_UNIT_AVG_WEEKLY_SALES')
plt.savefig('outliers1.png')
```

Identificar posibles outliers y anomalías

```python
#Se crea grafica boxplot por región
fig, ax = plt.subplots()
sns.boxplot(data = df, y = 'REGION', x = 'TOTAL_UNIT_AVG_WEEKLY_SALES')
plt.savefig('outliers2.png')
plt.show()
```

```python
#Se crea grafica boxplot por producto
fig, ax = plt.subplots()
sns.boxplot(data = df, y = 'ITEM', x = 'TOTAL_UNIT_AVG_WEEKLY_SALES')
ax.axes.yaxis.set_ticks([])
plt.savefig('outliers3.png')
plt.show()
```
